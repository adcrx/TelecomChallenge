```python
#Errores de formato e inconsistencia

#desanidamos las columnas y las convertimos en su propio df
df_customer = df['customer'].apply(pd.Series)
df_phone = df['phone'].apply(pd.Series)
df_internet = df['internet'].apply(pd.Series)
df_account = df['account'].apply(pd.Series)
```

```python
#Acá este df tiene más diccionarios
unique_account_simple = {col: df_account[col].unique() for col in ['Contract', 'PaperlessBilling', 'PaymentMethod']}
df_account_charges = df_account['Charges'].apply(pd.Series)
unique_charges = {col: df_account_charges[col].unique() for col in df_account_charges.columns}
unique_account_simple, unique_charges
```

###Manejo de inconsistencias

```python
#aplicamos los métodos de lower y strip() a todas las columnas de los nuevos df
df_customer = df_customer.apply(lambda col: col.str.lower().str.strip() if col.dtype == 'object' else col)
df_phone = df_phone.apply(lambda col: col.str.lower().str.strip() if col.dtype == 'object' else col)
df_internet = df_internet.apply(lambda col: col.str.lower().str.strip() if col.dtype == 'object' else col)
df_account_simple = df_account[['Contract', 'PaperlessBilling', 'PaymentMethod']].apply(lambda col: col.str.lower().str.strip() if col.dtype == 'object' else col)
```

###Creación columna cuentas diarias



```python
#Creamos la columna en base a la columna monthly
df_account_charges['Cuentas_Diarias'] = df_account_charges['Monthly'] / 30
df_account_charges[['Monthly', 'Cuentas_Diarias']].head()
```

###Estandarización y transformación de datos

```python
#Mapeo para convertir yes y no a 1 y 0
#!!!!!!!!!!!!!!!!!!!!! Esto terminó siendo pésima idea a la hora de graficar, así que la próxima vez se hará con mejor discreción
mapeo_binario = {'yes': 1, 'no': 0}
df_customer['Partner'] = df_customer['Partner'].map(mapeo_binario)
df_customer['Dependents'] = df_customer['Dependents'].map(mapeo_binario)

for col in df_customer.columns:
    if df_customer[col].dtype == 'object':
        df_customer[col] = df_customer[col].map(mapeo_binario).fillna(df_customer[col])

for col in df_phone.columns:
    if df_phone[col].dtype == 'object':
        df_phone[col] = df_phone[col].map(mapeo_binario).fillna(df_phone[col])

for col in df_internet.columns:
    if df_internet[col].dtype == 'object':
        df_internet[col] = df_internet[col].map(mapeo_binario).fillna(df_internet[col])

for col in ['Contract', 'PaperlessBilling', 'PaymentMethod']:
    df_account[col] = df_account[col].map(mapeo_binario).fillna(df_account[col])
```

##Paso 3: Carga y análisis

###Análisis descriptivo (media, mediana, desviación estándar)

```python
#Todos los df
dfs = {
    'df_customer': df_customer,
    'df_phone': df_phone,
    'df_internet': df_internet,
    'df_account': df_account,
    'df_account_charges': df_account_charges
    }

#diccionarios para los resultados
media = {}
mediana = {}
desviacionestandar = {}

for nombre_df, df_sub in dfs.items():
    #Selecciona todas las columnas numéricas
    columnas_numericas = df_sub.select_dtypes(include=['int64', 'float64']).columns

    media[nombre_df] = df_sub[columnas_numericas].mean()

    mediana[nombre_df] = df_sub[columnas_numericas].median()

    desviacionestandar[nombre_df] = df_sub[columnas_numericas].std()
```

###Recomendaciones

- Enfocar esfuerzos en clientes jóvenes y nuevos: Implementar programas de fidelización y beneficios especiales para clientes con menor tiempo de permanencia (tenure) y menores edades, ya que son los más propensos a cancelar.

- Promover contratos de mayor duración: Incentivar a los clientes para que opten por contratos anuales o bienales, ofreciendo descuentos o beneficios exclusivos, ya que los contratos mensuales presentan tasas de evasión más altas.

- Mejorar la oferta y comunicación de servicios adicionales: Potenciar la seguridad en línea, backup online, protección de dispositivos y soporte técnico como diferenciadores clave para aumentar la retención, destacando su valor y beneficios en la comunicación con el cliente.

- Segmentar la estrategia según tipo de servicio de Internet: Analizar más a fondo la mayor evasión en clientes con fibra óptica para identificar causas específicas y diseñar acciones específicas para este segmento (mejoras técnicas, ofertas exclusivas, atención personalizada).

- Optimizar métodos de pago: Incentivar el uso de métodos de pago automáticos (transferencia bancaria y tarjetas de crédito) frente a métodos con mayores tasas de evasión, como el cheque electrónico.

- Ofrecer alternativas para clientes con facturación electrónica: Revisar la experiencia y beneficios asociados a la facturación electrónica, mejorando el servicio para disminuir la tasa de cancelación en este grupo.

- Atender necesidades específicas de clientes sin pareja o sin personas a cargo: Evaluar productos o planes que puedan ajustarse mejor a las necesidades de estos segmentos, ofreciendo valor agregado para aumentar su permanencia.

- Mejorar el seguimiento y atención a clientes con servicio telefónico activo: Como este grupo tiene mayor evasión, se recomienda fortalecer el soporte, promociones y comunicación personalizada para aumentar su satisfacción.

- Desarrollar campañas de retención específicas para clientes con gasto mensual alto pero bajo gasto total: Estos clientes pueden estar en proceso de cancelar pronto; se pueden diseñar ofertas personalizadas para incrementar su lealtad.

- Monitorear constantemente las métricas de evasión por variable demográfica y de servicio: Utilizar análisis periódicos para ajustar estrategias y detectar nuevos patrones o segmentos en riesgo de evasión.
